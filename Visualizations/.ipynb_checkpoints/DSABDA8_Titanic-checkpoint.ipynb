```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df= sns.load_dataset('titanic')
df.head(5) 
```

```python
df.notnull()
```

```python
df.columns
```

## ScatterPlot

```python
sns.scatterplot(x='age',y='fare',data=df)
```

```python
sns.scatterplot(x='age',y='fare',hue = 'sex',data=df)
```

__Inference:__ Female passengers have less fare than male passengers

## Histogram

```python
sns.histplot(data=df, x='age')
```

__Inference:__ Maximum passengers were in the age group 20-40

```python
sns.histplot(data=df, x='fare')
```

# Displot

```python
sns.displot(df, x='age',bins=10)
```

```python
sns.displot(df, x='age',hue='sex', col='pclass')
```

*Inference: For each class of seats, the age group of people and their sex is visualized.

```python
sns.displot(df, x='age',hue='sex', col='embark_town')
```

*Inference: No. of people from each town of every age group differenciated by sex

```python
sns.displot(df, x='age',hue='survived', col='pclass')
```

*Inference: No. of people from each class of every age group differenciated by no. of people survived/dead

```python
sns.displot(df, x='age',hue='survived', col='embark_town')
```

Inference: No. of people from each town of every age group differenciated by no. of people survived/dead
